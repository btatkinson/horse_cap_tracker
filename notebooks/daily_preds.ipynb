```python

picks = pd.read_csv('C:\\Users\Blake\Downloads\Capping - CD-05-06.csv')
def format_pick_dict(series):
    
    cap = {}
    for i, row in enumerate(series):
        race_no = i+1
        cap_picks = row.split(',')
        cap_picks = [int(cp) for cp in cap_picks]
        top = cap_picks[0]
        a = cap_picks[1]
        b = cap_picks[2]
        cap[race_no] = {
            'Best':[],
            'Top':[top],
            'A':[a],
            'B':[b]
        }
    
    return cap

peck = format_pick_dict(picks['K. Peck'])
# sweep = format_pick_dict(picks['SWEEP'])
mcgee = format_pick_dict(picks['M. McGee'])
drf = format_pick_dict(picks['DRF Consensus'])
hersh = format_pick_dict(picks['M. Hersh'])

# hersh[11]['Best'] = 7
# hersh[11]['Top'] = []
```

```python

def convert_somich(path):
    df = pd.read_csv(path) ### CHECK HEADER
    df.columns=['Race','Top Pick','A','B']
    qs_picks = {}
    for index, row in df.iterrows():
        race = row['Race']
#         if 'R' in race:
#             race = race.replace('R','')
#             race = int(race)
        qs_picks[race] = {}
        qs_picks[race]['Best'] = []
        if not pd.isnull(row['Top Pick']):
            qs_picks[race]['Top'] = [int(row['Top Pick'])]
        else:
            qs_picks[race]['Top'] = []
        
        if pd.isnull(row['A']):
            A = []
        elif type(row['A'])==float:
            A = [row['A']]
        else:
            A = row['A'].split(',')
            A = [int(a) for a in A]
            
        if pd.isnull(row['B']):
            B = []
        elif type(row['B'])==float:
            B = [row['B']]
        else:
            B = row['B'].split(',')
            B = [int(b) for b in B]
        qs_picks[race]['A'] = A
        qs_picks[race]['B'] = B
    
    return qs_picks

ext = "Capping - Somich 05-06"
somich_path = f"C://Users/Blake/Downloads/{ext}.csv"
somich = convert_somich(somich_path)
```

```python

def convert_holmes(path):
    holmes = pd.read_csv(path) ### CHECK HEADER
    holmes.columns=['Race','Top Pick','A','B']
    qs_picks = {}
    for index, row in holmes.iterrows():
        race = row['Race']
        if 'R' in race:
            race = race.replace('R','')
            race = int(race)
        qs_picks[race] = {}
        qs_picks[race]['Best'] = []
        if not pd.isnull(row['Top Pick']):
            qs_picks[race]['Top'] = [int(row['Top Pick'])]
        else:
            qs_picks[race]['Top'] = []
        
        if pd.isnull(row['A']):
            A = []
        elif type(row['A'])==float:
            A = [row['A']]
        else:
            A = row['A'].split(',')
            A = [int(a) for a in A]
            
        if pd.isnull(row['B']):
            B = []
        elif type(row['B'])==float:
            B = [row['B']]
        else:
            B = row['B'].split(',')
            B = [int(b) for b in B]
        qs_picks[race]['A'] = A
        qs_picks[race]['B'] = B
    
    return qs_picks

ext = "Capping - Holmes 05-06"
holmes_path = f"C://Users/Blake/Downloads/{ext}.csv"
holmes = convert_holmes(holmes_path)
holmes
```

```python

def convert_jason(path):
    qs = pd.read_csv(path)
    qs_picks = {}
    for index, row in qs.iterrows():
        race = row['Race']
        qs_picks[race] = {}
        
        if not pd.isnull(row['Best']):
            qs_picks[race]['Best'] = [int(row['Best'])]
        else:
            qs_picks[race]['Best'] = []
            
        if not pd.isnull(row['Top']):
            qs_picks[race]['Top'] = [int(row['Top'])]
        else:
            qs_picks[race]['Top'] = []
        
        if pd.isnull(row['A']):
            A = []
        elif type(row['A'])==float:
            A = [row['A']]
        else:
            A = row['A'].split(',')
            A = [int(a) for a in A]
            
        if pd.isnull(row['B']):
            B = []
        elif type(row['B'])==float:
            B = [row['B']]
        else:
            B = row['B'].split(',')
            B = [int(b) for b in B if b.isnumeric()]
        qs_picks[race]['A'] = A
        qs_picks[race]['B'] = B
    
    return qs_picks

ext = "Daily Picks - 5_6 DERBY"
jason_path = f"C://Users/Blake/Downloads/{ext}.csv"
jason_o = convert_jason(jason_path)
```

```python

def convert_to_float(frac_str):
    try:
        return float(frac_str)
    except ValueError:
        num, denom = frac_str.split('/')
        try:
            leading, num = num.split(' ')
            whole = float(leading)
        except ValueError:
            whole = 0
        frac = float(num) / float(denom)
        return whole - frac if whole < 0 else whole + frac

def load_upc(track):
    entries = pd.read_csv(f'../data/upc_entries/{track}.csv')
    meta = pd.read_csv(f'../data/upc_meta/{track}.csv')
    entries['running'] = np.where(entries['horse_name'].isin([
        'Fogo Island','Peacock Lass', 'Be Here Now','T Max',
        'Price Talk','Legal Attack', 'Koufax','Richiesgotgame','Button Boy','Creative Minister',
        "Johny's Fireball","Flags Up","New Roo","Tap Dancing Lady","Stepper","Mouffy","Mambo Kitten",
        "Giant Game","Henrietta Topham","Kate's Kingdom","Effortlesslyelgant","Positano Sunset"
        
    ]), 0, entries['running'].copy())
    
    entries['ml_odds'] = entries['ml_odds'].apply(convert_to_float)
    entries = entries.loc[entries['running']==1].reset_index(drop=True)
    entries['implied_odds'] = entries['ml_odds'].copy().apply(lambda x: 1/(1+x))
    entries['implied_sum'] = entries.groupby(['date','track','race'])['implied_odds'].transform('sum')
    entries['implied_odds'] = entries['implied_odds'].copy()/entries['implied_sum'].copy()
    entries = entries.drop(columns=['implied_sum'])
    entries['race'] = entries['race'].astype(int)
#     for col in ['win_paid','place_paid','show_paid']:
#         entries[col] = entries[col].replace('[\$,]', '', regex=True).astype(float)
    
    return entries, meta



entries, meta = load_upc('churchill-downs')

```
